# 設置必要的套件
導入所需的視覺化套件，包括 Matplotlib、Seaborn 等，並設定基本樣式。

```python
# 導入所需的視覺化套件，包括 Matplotlib、Seaborn 等，並設定基本樣式
import matplotlib.pyplot as plt
import seaborn as sns

# 設定 Seaborn 的基本樣式
sns.set(style="whitegrid")

# 設定 Matplotlib 的基本樣式
plt.style.use('ggplot')
```

# 線性圖表繪製
使用 Matplotlib 繪製基本線圖，包括單線和多線圖表，並設定軸標籤、標題等。

```python
# 線性圖表繪製

# 使用 Matplotlib 繪製基本線圖，包括單線和多線圖表，並設定軸標籤、標題等。

# 繪製單線圖表
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o', linestyle='-', color='b', label='單線圖')
plt.xlabel('X 軸標籤')
plt.ylabel('Y 軸標籤')
plt.title('基本單線圖')
plt.legend()
plt.show()

# 繪製多線圖表
y2 = [1, 4, 6, 8, 10]

plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o', linestyle='-', color='b', label='線 1')
plt.plot(x, y2, marker='s', linestyle='--', color='r', label='線 2')
plt.xlabel('X 軸標籤')
plt.ylabel('Y 軸標籤')
plt.title('基本多線圖')
plt.legend()
plt.show()
```

# 散點圖製作
利用 Matplotlib 和 Seaborn 創建散點圖，展示數據點之間的關係。

```python
# 散點圖製作

# 創建一些示例數據
import numpy as np
np.random.seed(42)  # 設定隨機種子以確保結果可重現
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.rand(50) * 1000  # 點的大小
colors = np.random.rand(50)  # 點的顏色

# 使用 Matplotlib 繪製散點圖
plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=sizes, c=colors, alpha=0.5, cmap='viridis')
plt.colorbar()  # 顯示顏色條
plt.xlabel('X 軸標籤')
plt.ylabel('Y 軸標籤')
plt.title('Matplotlib 散點圖')
plt.show()

# 使用 Seaborn 繪製散點圖
tips = sns.load_dataset("tips")  # 加載示例數據集
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='day', size='size', palette='viridis', sizes=(20, 200))
plt.xlabel('總賬單')
plt.ylabel('小費')
plt.title('Seaborn 散點圖')
plt.legend()
plt.show()
```

# 長條圖呈現
製作垂直和水平長條圖，包括分組長條圖和堆疊長條圖。

```python
# 長條圖呈現

# 創建一些示例數據
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]

# 繪製垂直長條圖
plt.figure(figsize=(10, 6))
plt.bar(categories, values, color='skyblue')
plt.xlabel('類別')
plt.ylabel('值')
plt.title('垂直長條圖')
plt.show()

# 繪製水平長條圖
plt.figure(figsize=(10, 6))
plt.barh(categories, values, color='salmon')
plt.xlabel('值')
plt.ylabel('類別')
plt.title('水平長條圖')
plt.show()

# 創建分組長條圖的示例數據
values1 = [5, 7, 3, 8, 6]
values2 = [6, 9, 5, 7, 4]
bar_width = 0.35
index = np.arange(len(categories))

# 繪製分組長條圖
plt.figure(figsize=(10, 6))
plt.bar(index, values1, bar_width, label='組 1', color='skyblue')
plt.bar(index + bar_width, values2, bar_width, label='組 2', color='salmon')
plt.xlabel('類別')
plt.ylabel('值')
plt.title('分組長條圖')
plt.xticks(index + bar_width / 2, categories)
plt.legend()
plt.show()

# 創建堆疊長條圖的示例數據
values3 = [2, 3, 4, 1, 2]

# 繪製堆疊長條圖
plt.figure(figsize=(10, 6))
plt.bar(categories, values1, label='組 1', color='skyblue')
plt.bar(categories, values2, bottom=values1, label='組 2', color='salmon')
plt.bar(categories, values3, bottom=np.array(values1) + np.array(values2), label='組 3', color='lightgreen')
plt.xlabel('類別')
plt.ylabel('值')
plt.title('堆疊長條圖')
plt.legend()
plt.show()
```

# 圓餅圖展示
建立基本圓餅圖和甜甜圈圖，展示比例數據。

```python
# 圓餅圖展示

# 創建一些示例數據
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # 將第一塊分離出來

# 繪製基本圓餅圖
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('基本圓餅圖')
plt.axis('equal')  # 確保圓餅圖是圓的
plt.show()

# 繪製甜甜圈圖
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
# 在圓餅圖中間添加一個圓圈，使其成為甜甜圈圖
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('甜甜圈圖')
plt.axis('equal')  # 確保甜甜圈圖是圓的
plt.show()
```

# 熱力圖建立
使用 Seaborn 創建熱力圖，展示數據相關性矩陣。

```python
# 熱力圖建立

# 創建一些示例數據
data = np.random.rand(10, 12)

# 使用 Seaborn 創建熱力圖
plt.figure(figsize=(12, 8))
sns.heatmap(data, annot=True, fmt=".2f", cmap='viridis')
plt.title('數據相關性矩陣的熱力圖')
plt.show()
```
